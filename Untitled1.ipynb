# Top 5 Talkers

```python
top_talker = df_data['src_IP'].value_counts()
top5_talker=top_talker.head(5)
top5_talker
```

```python
import requests as req

for ip in top5_talker.index:
    result = req.get("http://ip-api.com/json/"+ip).json()["org"]
    print(f"IP : {ip} | Packet : {top5_talker[ip]} | Org : {result}")
```

# Exercise 4E

```python
group_sort = df_data.groupby(['src_IP','dst_IP']).size().sort_values(ascending=False)
top5_pair = group_sort.head(5)
print("Top 5 Pair")
top5_pair.head()
```

```python
import time
print("Top 5 pair IP and Org")
for source_IP, destination_IP in top5_pair.index:
    source_org = req.get("http://ip-api.com/json/"+source_IP).json()["org"]
    if source_org == "":
        source_org = "Returned EMPTY String"
    destination_org = req.get("http://ip-api.com/json/"+destination_IP).json()["org"]
    print(f"Source : {source_IP} : {source_org} || Destination : {destination_IP} : {destination_org}")
    time.sleep(0.1)
```
